```python
# Download model ZIP from github
!wget -O ML_Models.zip "https://github.com/mdalbakiakon/ML_Models/archive/refs/heads/main.zip"
!unzip -q ML_Models.zip

import tensorflow as tf
from PIL import Image
import requests
import numpy as np

# Load model from the unzipped folder
model = tf.keras.models.load_model('ML_Models-main/cats_vs_dogs_finetuned.keras')
```

```python
# Load image from URL
img_url = "https://www.whiskas.co.uk/sites/g/files/fnmzdf8046/files/2024-06/why-do-cats.jpg"
img = Image.open(requests.get(img_url, stream=True).raw).resize((320, 320))

# Preprocess
img_array = tf.keras.preprocessing.image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0) / 255.0

# Predict
pred = model.predict(img_array)

# Threshold
TH = 0.5
pred_class = int(pred[0][0] > TH)

# Define mapping manually
classes = {0: "cat", 1: "dog"}

print("Prediction:", classes[pred_class])
```
